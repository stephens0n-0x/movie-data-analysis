```python
x = np.random.rand(5000)
y = np.random.rand(5000)
inside_circle = (x - 0.5)**2 + (y - 0.5)**2 < 0.4**2
plt.scatter(x[inside_circle], y[inside_circle], c='g', marker='x')
plt.scatter(x[~inside_circle], y[~inside_circle], c='r', marker='o')
circle = plt.Circle((0.5, 0.5), 0.4, color='k', fill=False)
plt.gca().add_artist(circle)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```

```python
x = np.random.rand(5000)
y = np.random.rand(5000) * (1 - x)
dist = np.sqrt((x - 0.7)**2 + (y - 0.3)**2)
cmap = plt.get_cmap('cool')
plt.scatter(x, y, c=dist, cmap=cmap)
plt.xlim(0, 1)
plt.ylim(0, 1)
```
